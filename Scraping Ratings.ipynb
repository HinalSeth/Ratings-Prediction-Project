```python
# Importing Libraries
import selenium
import pandas as pd
import time

# Importing selenium webdriver 
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By

#Importing requests
import requests

# importing regex
import re

# Activating the chrome browser
s=Service("C:\\Users\\Hinal\\Downloads\\chromedriver_win32\\chromedriver.exe")  #opening chrome
driver = webdriver.Chrome(service=s)    
time.sleep(3)

# Opening the homepage of Amazon.in
url = "https://www.amazon.in/"
driver.get(url)

time.sleep(2)

product_search = driver.find_element(By.ID, "twotabsearchtextbox")
product_search.send_keys("Books")
driver.find_element(By.ID,'nav-search-submit-button').click()

URL=[]
Rev=[]
Rate=[]

for k in range(15):
    url = driver.find_elements(By.XPATH, "//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a")
    for i in url[:]:
        URL.append(i.get_attribute('href'))
    time.sleep(1)
    
    try:
        next_button = driver.find_element(By.XPATH,"//a[@class='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator']")
        next_button.click()
    except:
        break
```

```python
for i in URL:
    driver.get(i)   #fetching data from Each Product's URL
    time.sleep(1)
    
    #collecting 5 star reviews
    try:
        driver.find_element(By.XPATH, "//a[@class='a-link-normal 5star']").click()
        
        rv=driver.find_elements(By.XPATH,"//span[@class='a-size-base review-text review-text-content']")
        for j in rv:
            Rev.append(j.text)
            Rate.append('5')
    except:
        continue
        
     #collecting 4 star reviews
    try:
        driver.find_element(By.XPATH, "//a[@class='a-link-normal 4star']").click()
        
        rv=driver.find_elements(By.XPATH,"//span[@class='a-size-base review-text review-text-content']")
        for j in rv:
            Rev.append(j.text)
            Rate.append('4') 
    except:
        continue
        
    #collecting 3 star reviews
    try:
        driver.find_element(By.XPATH, "//a[@class='a-link-normal 3star']").click()
        
        rv=driver.find_elements(By.XPATH,"//span[@class='a-size-base review-text review-text-content']")
        for j in rv:
            Rev.append(j.text)
            Rate.append('3') 
    except:
        continue
        
    #collecting 2 star reviews
    try:
        driver.find_element(By.XPATH, "//a[@class='a-link-normal 2star']").click()
        
        rv=driver.find_elements(By.XPATH,"//span[@class='a-size-base review-text review-text-content']")
        for j in rv:
            Rev.append(j.text)
            Rate.append('2') 
    except:
        continue
        
    #collecting 1 star reviews
    try:
        driver.find_element(By.XPATH, "//a[@class='a-link-normal 1star']").click()
        
        rv=driver.find_elements(By.XPATH,"//span[@class='a-size-base review-text review-text-content']")
        for j in rv:
            Rev.append(j.text)
            Rate.append('1')
    except:
        continue
        
print(len(Rev),len(Rate))     
```

```python
#scraping data from flipkart
driver.get("https://www.flipkart.com/")
searchbar=driver.find_element(By.XPATH, "//div[@class='_3OO5Xc']/input")
searchbar.send_keys("Fashion")
driver.find_element(By.XPATH,"//button[@class='L0Z3Pu']").click()
URL3=[]
for k in range(10):
    url3 = driver.find_elements(By.XPATH, "//div[@class='_2B099V']")
    for i in url3[:]:
        URL3.append(i.get_attribute('href'))
    time.sleep(1)
    try:
        #click on the next button
        driver,find_element(By.XPATH, "//a[@class='_1LKTO3'][2]").click()
    except:
        break

driver.get("https://www.flipkart.com/")
searchbar=driver.find_element(By.XPATH, "//div[@class='_3OO5Xc']/input")
searchbar.send_keys("Jewelry")
driver.find_element(By.XPATH,"//button[@class='L0Z3Pu']").click()    
for k in range(10):
    url3 = driver.find_elements(By.XPATH, "//div[@class='_2B099V']")
    for i in url3[:]:
        URL3.append(i.get_attribute('href'))
    time.sleep(1)
    try:
        #click on the next button
        driver,find_element(By.XPATH, "//a[@class='_1LKTO3'][2]").click()
    except:
        break

len(URL3)
```

```python
for i in URL3:
    driver.get(i)   #fetching data from Each Product's URL
    time.sleep(1)
    
    #collecting all reviews
    try:
        rv=driver.find_elements(By.XPATH,"//div[@class='_6K-7Co']")
        for j in rv:
            Rev.append(j.text)
    except:
        Rev.append('No Review available')
    try:
        rt=driver.find_elements(By.XPATH,"//div[@class='_3LWZlK _1BLPMq _3B8WaH']")
        for j in rt:
            rate=rt.text
            rate=rate.split(' ')[0]
            Rate.append(rate)
    except:
        pass
        
print(len(Rev),len(Rate))     
```

```python
#creating dataframe and saving csv file
dataset=pd.DataFrame({'Rating':Rate,'Review':Rev})
dataset.to_csv("Dataset_Ratings.csv",index=False)
```
